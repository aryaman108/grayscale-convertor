```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
input_image=imread("rgb_img.png")

r,g,b=input_image[:,:,0],input_image[:,:,1],input_image[:,:,2]

#this can be changed
gamma=1

#this cannot be changed
r_const,g_const,b_const=0.2989,0.5870,0.1140

# Calculate the grayscale image using the corrected channels
grayscale_image = (r_const * r) ** gamma + (g_const * g) ** gamma + (b_const * b) ** gamma

fig=plt.figure(1)

img1,img2=fig.add_subplot(121),fig.add_subplot(122)

img1.imshow(input_image)
img2.imshow(grayscale_image,cmap="gray")

fig.show()
plt.show()
```
